```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('D:\Courses\Machine Learning Andrew NG\Week 2\machine-learning-ex1\ex1\ex1data1.txt',delimiter=',')
X = data[:,0]
y = data[:,1]
```

# Plotting data

```python
plt.scatter(X,y,s=30,c='r',marker='x',linewidths=1)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s');
plt.show()
```

# Cost Function

```python
def costFunction(X,y,theta=[0,0]):
    x = np.c_[np.ones(X.shape[0]),X]
    m = len(y);
    J = 0;
    prediction = x.dot(theta)
    sqrErrors = np.square((prediction-y));
    J = 1/(2*m) * np.sum(sqrErrors);
    return J
```

# Gradient Function

```python
def gradientDescent(X,y,theta=[0,0],alpha=.01,num_iters=1500):
    m=len(y)
    J_history = np.zeros((num_iters,1))
    
    for iter in np.arange(num_iters):
        h = theta[0]+ (theta[1]*X);
        theta_zero = theta[0] - alpha * (1/m) * np.sum(h-y)
        theta_one = theta[1] - alpha * (1/m) * np.sum(np.multiply((h-y),X))
        theta = [theta_zero,theta_one];
        J_history[iter] = costFunction(X,y,theta)
    return (theta,J_history)
```

```python
theta,J_history = gradientDescent(X,y)
```

# Plotting linear line in graph

```python
plt.plot(J_history)
plt.xlabel('Iterations')
plt.ylabel('Cost Function')
plt.show()
```

# Plot the linear regression line

```python
x = np.c_[np.ones(X.shape[0]),X]
xx= np.arange(5,23)
yy = theta[0]+theta[1]*xx
plt.scatter(X,y,s=30,c='r',marker='x',linewidths=1)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s');
plt.plot(xx,yy,'-')
plt.show()
```
